```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/C34B342-1522.csv',delimiter=";")
filtroC = ['C340','C341','C342','C343','C348','C349']
filtroB = ['B342']
dfC34 = df[df['DIAG_PRINC'].isin(filtroC)]
dfB34 = df[df['DIAG_PRINC'].isin(filtroB)]
```

```python
dfB34.groupby(['ANO_CMPT', 'MES_CMPT']).size().plot(label='B34')
```

```python
dfC34.groupby(['ANO_CMPT', 'MES_CMPT']).size().plot(label='C34')
```

```python
data1 = dfC34.groupby(['ANO_CMPT', 'MES_CMPT']).size().unstack(fill_value=0).stack().values
data2 = dfB34.groupby(['ANO_CMPT', 'MES_CMPT']).size().unstack(fill_value=0).stack()
data2
```

```python

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('GDP Country X (trillion)', color=color)

ax1.plot(data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:orange'
ax2.set_ylabel('GDP City Y (million)', color=color) 
ax2.plot(data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout() 
plt.legend()
plt.show()
```
